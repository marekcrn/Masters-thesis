```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import numpy as np
```

```python
RX_pattern = np.load('data/RX_pattern.npy')
TX_pattern = np.load('data/TX_pattern.npy')
OMNI_pattern = np.load('data/OMNI_pattern.npy')
impulse_response = np.load('data/impulse_response.npy')
#impulse_response = np.load('data/impulse_response_NLoS.npy')
#print(impulse_response)
```

```python
def horzontal_shift(angle, pattern): 
    if angle > 0:
        horizontal1 = pattern[:,-angle:]
        horizontal2 = pattern[:,:-angle]
        shifted_pattern = np.hstack((horizontal1, horizontal2))
        
    elif angle < 0:
        horizontal1 = pattern[:,-(360+angle):]
        horizontal2 = pattern[:,:-(360+angle)]
        shifted_pattern = np.hstack((horizontal1, horizontal2))
        
    else:
        shifted_pattern = pattern  

    return shifted_pattern
```

```python
def vertical_shift(angle, pattern): 
    if angle > 0:
        vertical1 = pattern[-angle:,:]
        vertical2 = pattern[:-angle,:]
        shifted_pattern = np.vstack((vertical1, vertical2))
        
    elif angle < 0:
        vertical1 = pattern[-(360+angle):,:]
        vertical2 = pattern[:-(360+angle),:]
        shifted_pattern = np.vstack((vertical1, vertical2))
        
    else:
        shifted_pattern = pattern  

    return shifted_pattern
```

```python
all_paths = impulse_response.shape[0]
#TX_angles = np.linspace(-45, 45, num=19) 
#TX_angles = np.roll(TX_angles, -9) 
TX_angles = np.concatenate((np.linspace(-45, -1, num=45),[0], np.linspace(1, 45, num=45)))
TX_angles = np.roll(TX_angles, -45)
TX_sweep = np.zeros((all_paths, len(TX_angles)+1))
a = np.zeros((all_paths, len(TX_angles)))
i = 0

for angle in TX_angles:
    TX_pattern_shifted = horzontal_shift(int(angle),TX_pattern)
    
    for path_idx in range(all_paths):
        TX1 = int(impulse_response[path_idx][3])
        TX2 = int(impulse_response[path_idx][4])
        RX1 = int(impulse_response[path_idx][5])
        RX2 = int(impulse_response[path_idx][6])

        TX_sweep[path_idx][0] = impulse_response[path_idx][2]
        TX_sweep[path_idx][(i+1)] = 20*np.log10(impulse_response[path_idx][1])
        TX_sweep[path_idx][(i+1)] = TX_sweep[path_idx][(i+1)] + TX_pattern_shifted[TX1][TX2] + OMNI_pattern[RX1][RX2]
        TX_sweep[path_idx][(i+1)] = 10**(TX_sweep[path_idx][(i+1)]/20)

        a[path_idx][i] = TX_pattern_shifted[TX1][TX2]
        
    i = i + 1
    
#print(TX_sweep)
```

```python
def angles_shift(angle): 
    if angle > 0:
        angle_axis = np.concatenate((angle_vector[-(360-angle):], angle_vector[:-(360-angle)])) 
    elif angle < 0:
        angle_axis = np.concatenate((angle_vector[-(-angle):], angle_vector[:-(-angle)]))
    else:
        angle_axis = angle_vector

    return angle_axis
```

```python
t = TX_sweep[:,0]/1e-9
angle_vector = np.linspace(0, 359, num=360)*np.pi/180
max_a_TX = np.max(TX_sweep)
min_a_TX = -0.01*np.average(TX_sweep)
i=0

for angle in TX_angles: 
    angle_axis = angles_shift(int(angle))  
  
    # CIR time domain
    plt.figure()
    plt.title("Channel impulse response realization for " + str(int(angle)) + "°", fontsize=15)
    plt.stem(t, TX_sweep[:,(i+1)], linefmt='blue')
    plt.xlim([0, np.max(t)])
    plt.ylim([-np.average(min_a_TX), max_a_TX*1.1])
    plt.xlabel(r"$\tau$ [ns]", fontsize=15)
    plt.ylabel(r"$|a|$", fontsize=15);
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.grid()

    # Save each plot with a different filename
    filename = f"images/Time/TX/CIR{i}.png"
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    plt.close()

    fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
    ax.scatter(impulse_response[:, 4] * np.pi / 180, a[:,i], color='blue', label='CIR')
    ax.plot(angle_axis, TX_pattern[90, :], color='red')
    ax.set_title("CIR - Azimuth angle of departure", fontsize=15)
    ax.set_ylabel('EIRP [dBm]', labelpad=30, fontsize=15)
    ax.tick_params(axis='y',labelsize=12)
    ax.tick_params(axis='x',labelsize=12)
    ax.grid(True)

    # Save each plot with a different filename
    filename = f"images/Angle/TX_sweep_TX/CIR{i}.png"
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    plt.close(fig)  # Close figure to prevent memory issues
    
    fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
    ax.scatter(impulse_response[:,6]*np.pi/180, TX_sweep[:,(i+1)], color='blue')
    ax.set_title("CIR - Azimuth angle of arrival", fontsize=15)
    ax.yaxis.set_major_locator(MaxNLocator(nbins=5))
    ax.set_ylim(np.abs(min_a_TX),max_a_TX*1.2)
    ax.tick_params(axis='y',labelsize=12)
    ax.tick_params(axis='x',labelsize=12)
    ax.grid(True)

    ax_secondary = fig.add_axes(ax.get_position(), projection='polar')
    ax_secondary.plot(angle_vector, OMNI_pattern[90, :], color='red')
    ax.set_ylabel(r"$|a|$", labelpad=30, fontsize=15)
    ax_secondary.set_frame_on(False)  # Remove the frame
    ax_secondary.yaxis.set_visible(False)  # Hide secondary axis ticks and labels
    ax_secondary.xaxis.set_visible(False)  # Optionally hide angular axis
    ax_secondary.set_ylim(-30, 5)

    # Save each plot with a different filename
    filename = f"images/Angle/TX_sweep_RX/CIR{i}.png"
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    plt.close(fig)  # Close figure to prevent memory issues

    i = i + 1
```

```python
TX_codebook = np.zeros(len(TX_angles))

for TX_angle_idx in range(len(TX_codebook)):
    TX_codebook[TX_angle_idx] = np.max(TX_sweep[:,(TX_angle_idx+1)])

max_amp_TX = np.max(TX_codebook)
max_idx_TX = np.argmax(TX_codebook)

print("Maximum amplitude is " + str(max_amp_TX) + " for TX angle " + str(TX_angles[max_idx_TX]) + "°")
```

```python
TX_pattern_shifted = horzontal_shift(TX_angles[max_idx_TX],TX_pattern)
#RX_angles_h = np.linspace(-45, 45, num=19)
#RX_angles_h = np.roll(RX_angles_h , -9) 
RX_angles_h = np.concatenate((np.linspace(-45, -1, num=45),[0], np.linspace(1, 45, num=45)))
RX_angles_h = np.roll(RX_angles_h , -45)
RX_angles_v = [-15,0,15]
RX_sweep = np.zeros((all_paths, len(RX_angles_h)+1, len(RX_angles_v)))
a2 = np.zeros((all_paths, len(RX_angles_h), len(RX_angles_v)))
k = 0

for  angle_v in RX_angles_v:
    i = 0 
    RX_pattern_shifted_v = vertical_shift(angle_v,RX_pattern)
    
    for angle_h in RX_angles_h:
        RX_pattern_shifted = horzontal_shift(int(angle_h),RX_pattern_shifted_v)
       
        for path_idx in range(all_paths):
            TX1 = int(impulse_response[path_idx][3])
            TX2 = int(impulse_response[path_idx][4])
            RX1 = int(impulse_response[path_idx][5])
            RX2 = int(impulse_response[path_idx][6])

            RX_sweep[path_idx][0][k] = impulse_response[path_idx][2]
            RX_sweep[path_idx][(i+1)][k] = 20*np.log10(impulse_response[path_idx][1])
            RX_sweep[path_idx][(i+1)][k] = RX_sweep[path_idx][(i+1)][k] + TX_pattern_shifted[TX1][TX2] + RX_pattern_shifted[RX1][RX2]
            RX_sweep[path_idx][(i+1)][k] = 10**(RX_sweep[path_idx][(i+1)][k]/20)

            a2[path_idx][i][k] = TX_pattern_shifted[TX1][TX2]

        i = i + 1
        
    k = k + 1
      
#print(RX_sweep)
```

```python
angle_axis_TX = angles_shift(TX_angles[max_idx_TX])
vertical_angle_position = 0 # 0=-15,1=0,2=15
vertical_angle_value = RX_angles_v[vertical_angle_position]
i=0

for angle in RX_angles_h: 
    angle_axis_RX = angles_shift(int(angle)) 
    
    max_a_RX = np.max(RX_sweep[:,:,vertical_angle_position])
    min_a_RX = -0.01*np.average(RX_sweep[:,:,vertical_angle_position])
    
    # CIR time domain
    plt.figure()
    plt.title("Channel impulse response realization for " + str(int(angle)) + "° horizontal and " + str(int(vertical_angle_value)) + "° vertical")
    plt.stem(t, RX_sweep[:,(i+1),vertical_angle_position], linefmt='blue')
    plt.xlim([0, np.max(t)])
    plt.ylim([min_a_RX,max_a_RX*1.1])
    plt.xlabel(r"$\tau$ [ns]", fontsize=15)
    plt.ylabel(r"$|a|$", fontsize=15);
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.grid()

    # Save each plot with a different filename
    filename = f"images/Time/RX/CIR{i}.png"
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    plt.close()
    
    # CIR angle domain
    fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
    ax.scatter(impulse_response[:, 4] * np.pi / 180, a2[:,i,vertical_angle_position], color='blue', label='CIR')
    ax.plot(angle_axis_TX, TX_pattern[90, :], color='red')
    ax.set_title("CIR - Azimuth angle of departure", fontsize=15)
    ax.set_ylabel('EIRP [dBm]', labelpad=30, fontsize=15)
    ax.tick_params(axis='y',labelsize=12)
    ax.tick_params(axis='x',labelsize=12)
    ax.grid(True)
    plt.close(fig)  # Close figure to prevent memory issues
 
    fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
    # Plot on the second subplot
    ax.scatter(impulse_response[:,6]*np.pi/180, RX_sweep[:,(i+1),vertical_angle_position], color='blue')
    ax.set_title("CIR - Azimuth angle of arrival", fontsize=15)
    ax.yaxis.set_major_locator(MaxNLocator(nbins=5))
    ax.set_ylim(np.abs(min_a_RX),max_a_TX*1.1)
    ax.tick_params(axis='y',labelsize=12)
    ax.tick_params(axis='x',labelsize=12)
    ax.grid(True)

    ax_secondary = fig.add_axes(ax.get_position(), projection='polar')
    ax_secondary.plot(angle_axis_RX, RX_pattern[90, :], color='red')
    ax.set_ylabel(r"$|a|$", labelpad=30, fontsize=15)
    ax_secondary.set_frame_on(False)  # Remove the frame
    ax_secondary.yaxis.set_visible(False)  # Hide secondary axis ticks and labels
    ax_secondary.xaxis.set_visible(False)  # Optionally hide angular axis
    ax_secondary.set_ylim(-60, 5)

    # Save each plot with a different filename
    filename = f"images/Angle/RX_sweep_RX/CIR{i}.png"
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    plt.close(fig)  # Close figure to prevent memory issues

    i = i + 1
```
