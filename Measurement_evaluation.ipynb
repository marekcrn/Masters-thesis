```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import numpy as np
```

```python
LoS_simulation = np.load('data/LoS_simulation.npy')
NLoS_simulation = np.load('data/NLoS_simulation.npy')
measurement_1 = np.load('data/measurement_20_2_2025.npy') # LoS, NLoS-omni, NLoS-directional (0 degree)
#print(measurement_1)
```

```python
sweep_angles = np.linspace(-45,45,60)
TX_gain = 52
RX_omni_gain = 0
RX_direct_gain = 16
```

```python
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.scatter((sweep_angles-42)*np.pi/180, measurement_1[:,0], color='red', label="measured")
ax.scatter(LoS_simulation[:,4]*np.pi/180, TX_gain+RX_omni_gain+10*np.log10(LoS_simulation[:,1]), color='blue', label="simulated")
ax.yaxis.set_major_locator(MaxNLocator(nbins=5))
angle = np.deg2rad(53)
ax.legend(loc="lower left", fontsize=12, bbox_to_anchor=(.5 + np.cos(angle)/2, .5 + np.sin(angle)/2))
ax.set_title("CIR - Azimuth angle of departure", fontsize=15)
ax.set_ylabel("RSS [dBm]", labelpad=30, fontsize=15)
ax.tick_params(axis='y',labelsize=12)
ax.tick_params(axis='x',labelsize=12)
ax.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.scatter((sweep_angles-115)*np.pi/180, measurement_1[:,1], color='red', label="measured")
ax.scatter(NLoS_simulation[:,4]*np.pi/180, TX_gain+RX_omni_gain+10*np.log10(NLoS_simulation[:,1]), color='blue', label="simulated")
ax.yaxis.set_major_locator(MaxNLocator(nbins=5))
angle = np.deg2rad(53)
ax.legend(loc="lower left", fontsize=12, bbox_to_anchor=(.5 + np.cos(angle)/2, .5 + np.sin(angle)/2))
ax.set_title("CIR - Azimuth angle of departure", fontsize=15)
ax.set_ylabel("RSS [dBm]", labelpad=30, fontsize=15)
ax.tick_params(axis='y',labelsize=12)
ax.tick_params(axis='x',labelsize=12)
ax.grid(True)
plt.show()
```
