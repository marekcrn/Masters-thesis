```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import numpy as np
from scipy.optimize import curve_fit
```

```python
# Simulations
LoS_simulation = np.load('data/LoS_simulation.npy')
NLoS_simulation = np.load('data/NLoS_simulation.npy')

# First measurement - Angle of departure
measurement_1 = np.load('data/measurement_20_2_2025.npy') # LoS, NLoS-omni, NLoS-directional (0 degree)

# Second measurement - Angle of arrival
measurement_2 = np.load('data/measurement_15_4_2025.npy') # LoS, NLoS-omni, NLoS-directional (0 degree)
BER_LOS = np.load('data/measured_ber_LOS.npy') 
BER_NLOS = np.load('data/measured_ber_NLOS.npy') 
```

```python
TX_EIRP_directional = 52
TX_EIRP_waveguide_LOS = 65
TX_EIRP_waveguide_NLOS = 85

RX_EIRP_omni = 0
RX_EIRP_directional = 16
```

# First measurement - Angle of departure

```python
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.scatter((measurement_2[0:60,0]-38)*np.pi/180, measurement_1[:,0], color='red', label="measured")
ax.scatter(LoS_simulation[:,4]*np.pi/180, TX_EIRP_directional+RX_EIRP_omni+10*np.log10(LoS_simulation[:,1]), color='blue', label="simulated")
ax.yaxis.set_major_locator(MaxNLocator(nbins=5))
angle = np.deg2rad(53)
ax.legend(loc="lower left", fontsize=12, bbox_to_anchor=(.5 + np.cos(angle)/2, .5 + np.sin(angle)/2))
ax.set_title("CIR - Azimuth angle of departure", fontsize=15)
ax.set_ylabel("RSS [dBm]", labelpad=30, fontsize=15)
ax.tick_params(axis='y',labelsize=12)
ax.tick_params(axis='x',labelsize=12)
ax.grid(True)
plt.show()
```

# Second measurement - Angle of arrival

```python
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.scatter((measurement_2[:,0]+147)*np.pi/180, measurement_2[:,1], color='red', label="measured")
ax.scatter(LoS_simulation[:,6]*np.pi/180,TX_EIRP_waveguide_LOS+RX_EIRP_directional+10*np.log10(LoS_simulation[:,1]), color='blue', label="simulated")
ax.yaxis.set_major_locator(MaxNLocator(nbins=5))
angle = np.deg2rad(53)
ax.legend(loc="lower left", fontsize=12, bbox_to_anchor=(.5 + np.cos(angle)/2, .5 + np.sin(angle)/2))
ax.set_title("CIR - Azimuth angle of arrival", fontsize=15)
ax.set_ylabel("RSS [dBm]", labelpad=30, fontsize=15)
ax.tick_params(axis='y',labelsize=12)
ax.tick_params(axis='x',labelsize=12)
ax.grid(True)
plt.show()
```

```python
snr_LOS_aprox = BER_LOS[0:53,0]
ber_LOS_aprox = BER_LOS[0:53,1]

snr_NLOS_aprox = BER_NLOS[:,0]
ber_NLOS_aprox = BER_NLOS[:,1]

# Setting zero or negative BER values to a small positive value
ber_LOS_aprox[ber_LOS_aprox <= 0] = 1e-6
ber_NLOS_aprox[ber_NLOS_aprox <= 0] = 1e-6

# Define a sigmoid (logistic) model for better fit
def sigmoid(x, a, b, c, d):
    return a / (1.0 + np.exp(-b * (x - c))) + d

# Initial guess for curve fitting
initial_guess_LOS = [1, -1, np.median(snr_LOS_aprox), 1e-5]
initial_guess_NLOS = [1, -1, np.median(snr_NLOS_aprox), 1e-5]

# Curve fitting with sigmoid model
popt_LOS, _ = curve_fit(sigmoid, snr_LOS_aprox, ber_LOS_aprox, p0=initial_guess_LOS, maxfev=10000)
popt_NLOS, _ = curve_fit(sigmoid, snr_NLOS_aprox, ber_NLOS_aprox, p0=initial_guess_NLOS, maxfev=10000)

# Generating smooth SNR values
snr_LOS_aprox = np.linspace(snr_LOS_aprox.min(), snr_LOS_aprox.max(), 500)
snr_NLOS_aprox = np.linspace(snr_NLOS_aprox.min(), snr_NLOS_aprox.max(), 500)

ber_LOS_aprox = sigmoid(snr_LOS_aprox, *popt_LOS)
ber_NLOS_aprox = sigmoid(snr_NLOS_aprox, *popt_NLOS)
```

```python
plt.figure()
plt.title("Symbol error rate for LOS scenario", fontsize=15)
plt.xlabel("SNR [dB]", fontsize=15)
plt.ylabel("SER [-]", fontsize=15)
plt.grid(which="both")
plt.ylim([1e-5, 1.0])
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.semilogy(BER_LOS[:,0], BER_LOS[:,1], color='blue')
plt.semilogy(snr_LOS_aprox,ber_LOS_aprox, color='red')
plt.legend(['Measured SER', 'Approximated SER'], fontsize=15, loc="lower left")
```

```python
plt.figure()
plt.title("Symbol error rate for NLOS scenario", fontsize=15)
plt.xlabel("SNR [dB]", fontsize=15)
plt.ylabel("SER [-]", fontsize=15)
plt.grid(which="both")
plt.ylim([1e-5, 1.0])
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.semilogy(BER_NLOS[:,0], BER_NLOS[:,1], color='blue')
plt.semilogy(snr_NLOS_aprox,ber_NLOS_aprox, color='red')
plt.legend(['Measured SER', 'Approximated SER'], fontsize=15, loc="lower left")
```
