```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import numpy as np
```

```python
LoS_simulation = np.load('data/LoS_simulation.npy')
NLoS_simulation = np.load('data/NLoS_simulation.npy')
measurement_1 = np.load('data/measurement_20_2_2025.npy') # LoS, NLoS-omni, NLoS-directional (0 degree)
#print(measurement_1)
```

```python
sweep_angles = np.linspace(-45,45,60)
```

```python
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})

ax.yaxis.set_major_locator(MaxNLocator(nbins=5))
ax.scatter((sweep_angles-42)*np.pi/180, measurement_1[:,0], color='red')
ax.yaxis.set_major_locator(MaxNLocator(nbins=5))

ax_secondary = fig.add_axes(ax.get_position(), projection='polar')
#ax_secondary.scatter(LoS_simulation[:,4]*np.pi/180, np.ones((LoS_simulation.shape[0], 1))*42, color='blue')
ax_secondary.scatter(LoS_simulation[:,4]*np.pi/180, 42+10*np.log10(LoS_simulation[:,1]), color='blue')


ax_secondary.set_frame_on(False)  # Remove the frame
ax_secondary.yaxis.set_visible(False)  # Hide secondary axis ticks and labels
ax_secondary.xaxis.set_visible(False)

ax.set_title("CIR")
ax.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})

ax.yaxis.set_major_locator(MaxNLocator(nbins=5))
ax.scatter((sweep_angles-115)*np.pi/180, measurement_1[:,1], color='red')
ax.yaxis.set_major_locator(MaxNLocator(nbins=5))

ax_secondary = fig.add_axes(ax.get_position(), projection='polar')
#ax_secondary.scatter(NLoS_simulation[:,4]*np.pi/180, np.ones((NLoS_simulation.shape[0], 1))*42, color='blue')
ax_secondary.scatter(NLoS_simulation[:,4]*np.pi/180, 42+10*np.log10(NLoS_simulation[:,1]), color='blue')

ax_secondary.set_frame_on(False)  # Remove the frame
ax_secondary.yaxis.set_visible(False)  # Hide secondary axis ticks and labels
ax_secondary.xaxis.set_visible(False)

ax.set_title("CIR")
ax.grid(True)
plt.show()
```
